```python
# Trim long video to a 30sec segment for testing
import cv2

input_path = "data/videos/2026_Bball.mp4"
output_path = "data/videos/2026_Bball_trimmed.mp4"

cap = cv2.VideoCapture(input_path)
fps = cap.get(cv2.CAP_PROP_FPS)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

start = 60
end = 90

start_frame = int(start * fps)
end_frame   = int(end * fps)

cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

for f in range(start_frame, end_frame):
    ret, frame = cap.read()
    if not ret:
        break
    out.write(frame)

cap.release()
out.release()
```

```python
# Run inference on the segmented part of the video
model = YOLO("runs/detect/train7/weights/best.pt")
results = model.predict("data/videos/2026_Bball_trimmed.mp4", save=True, classes=[0, 1])
```
