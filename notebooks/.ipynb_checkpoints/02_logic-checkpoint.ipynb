```python
# This notebook will clean up the detection for the basketball hoop, choosing only the largest one to use
# Track the ball
# Determine the hoops coordinates
# Build a state machine to determine when the ball is near the hoop and then will run a check to see if it passes through
```

```python
import sys
import os
os.chdir("C:/Users/matth/Documents/UNICODE/CompVis/compvisfinal_real/Comp_vis_Final")
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Assign confidence threshold
# This threshold is essential for stopping false positives from 
# affecting our interpolation
conf_thresh = 0.35
```

```python
def get_hoop_box(results, model):
    hoop_boxes = []

    # Loop through all detections
    for b in results.boxes:
        cls = int(b.cls[0])      # class index (0 = ball, 1 = rim)
        conf = float(b.conf[0])  # confidence score

        # Keep only rim detections above threshold
        if model.names[cls] == "Basketball Rim" and conf >= conf_thresh:
            hoop_boxes.append(b)

    # No valid rim detections
    if not hoop_boxes:
        return None

    # Pick the largest rim box (helps avoid false positives)
    hoop_boxes.sort(
        key=lambda b: (b.xyxy[0][2] - b.xyxy[0][0]) * (b.xyxy[0][3] - b.xyxy[0][1]),
        reverse=True
    )

    hoop = hoop_boxes[0]

    # Extract coordinates
    x1, y1, x2, y2 = hoop.xyxy[0]

    # Only keep the top band of the rim
    y2_new = y1 + 10

    return (float(x1), float(y1), float(x2), float(y2_new))
```

```python
def get_ball_center(results, model):
    ball_boxes = []

    # Loop through all detections
    for b in results.boxes:
        cls = int(b.cls[0])      # class index (0 = Basketball)
        conf = float(b.conf[0])  # confidence score

        # Keep only basketball detections above threshold
        if model.names[cls] == "Basketball" and conf >= conf_thresh:
            ball_boxes.append(b)

    if not ball_boxes:
        return None

    # Choose the highest-confidence ball
    ball = max(ball_boxes, key=lambda b: float(b.conf[0]))

    x1, y1, x2, y2 = ball.xyxy[0]
    cx = float((x1 + x2) / 2)
    cy = float((y1 + y2) / 2)

    return (cx, cy)
```

```python
def get_ball_zone(ball_center, hoop_box):
    bx, by = ball_center
    x1, y1, x2, y2 = hoop_box  # unpack tuple

    # y1 = top of rim band
    # y2 = bottom of rim band (y1 + thickness)

    if by < y1:
        return "above"
    elif y1 <= by <= y2 and x1 <= bx <= x2:
        return "inside"
    else:
        return "below"
```

```python
def update_state(state, zone):
    global made_shots

    state = "idle"

    if zone == "above":
        return "shot_started"
    elif zone == "below":
        return "idle"

    # Determing if ball went through hoop
    return state
```

```python
def did_ball_cross_hoop(hoop, x_smooth, y_smooth):
    if hoop is None or x_smooth is None or y_smooth is None:
        return False

    x1, y1, x2, y2 = hoop
    hx = (x1 + x2) / 2
    hy = y1 # top of rim band
    rim_radius = (x2 - x1) / 2

    for i in range(1, len(x_smooth)):
        x_prev, y_prev = x_smooth[i-1], y_smooth[i-1]
        x_curr, y_curr = x_smooth[i],   y_smooth[i]

        # Check vertical crossing
        crossed_vertical = (y_prev < hy) and (y_curr > hy)

        if not crossed_vertical:
            continue

        # Check horizontal alignment
        inside_horizontal = abs(x_curr - hx) < rim_radius

        if inside_horizontal:
            return True

    return False
```

```python
# Used to determine if the trajectory was near the hoop to see
# if the ball was actually an attempted shot
def trajectory_near_hoop(hoop, x_smooth, y_smooth, factor=1.5):
    if hoop is None or x_smooth is None:
        return False

    x1, y1, x2, y2 = hoop
    hx = (x1 + x2) / 2
    rim_radius = (x2 - x1) / 2

    # Allow for some tolerance
    allowed_dist = rim_radius * factor

    # Check if any point in the trajectory is horizontally NEAR the hoop
    for x in x_smooth:
        if abs(x - hx) < allowed_dist:
            return True

    return False
        
```

```python
# Assign velocity threshold
# A ball travles smoothly in a consistent motion, if it every jumps say 80 pixels
# or more in a millisecond, we can disregard that as a false read.

prev_ball = None
MAX_SPEED = 75 # px per frame

# Implement a counter for lost frames and max lost frames
# before we count the ball as "lost" to help solve errors with velocity tracking
lost_frames = 0
MAX_LOST = 20
```

```python
# Full inference loop to go through a video and track points
model = YOLO("runs/detect/train7/weights/best.pt") 

# =======Open Input Video and Set Output video, Time Crop=========
# Set the input and output paths
input_path = "data/videos/2026_Bball.mp4"
output_path = "output/Point_counting_results.mp4"

# Open Video
cap = cv2.VideoCapture(input_path)
fps = cap.get(cv2.CAP_PROP_FPS)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Time crop period in seconds
start = 0 
end = 120

start_frame = int(start * fps)
end_frame   = int(end * fps)

cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

# =======Initialize variables=========
# Use the state machine to figure out tracking of points
state = "idle"
made_shots = 0
attempted_shots = 0

# Create a list to track the trajectory of the center of ball through interpolation
trajectory = []   # list of (frame_idx, x, y)
x_smooth, y_smooth = None, None
recording = False
frame_idx = 0

cooldown = 0
# COOLDOWN_FRAMES is how many frames we record after ball goes in
COOLDOWN_FRAMES = 5

# =======Begin Looping Through Video=========
# Run a loop through every frame of the video
# 1. We need to interpolate the travelling of the ball
# 2. Get the bbox's of the ball and hoop
while True:
    current_frame = cap.get(cv2.CAP_PROP_POS_FRAMES)
    if current_frame >= end_frame:
        break

    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO
    results = model(frame, classes=[0, 1])[0]
    
    # Extract hoop + ball
    hoop = get_hoop_box(results, model)

    # Run velocity check when getting center of ball
    raw_ball = get_ball_center(results, model)
    ball_center = None

    if raw_ball:
        if prev_ball is None:
            ball_center = raw_ball
            lost_frames = 0
        else:
            # Calculate the difference in coordinates from the last ball seen and the one now
            dx = raw_ball[0] - prev_ball[0]
            dy = raw_ball[1] - prev_ball[1]
            dist = (dx*dx + dy*dy)**0.5
    
            if dist < MAX_SPEED:
                ball_center = raw_ball
                lost_frames = 0
            else:
                # reject this detection as a false positive
                ball_center = None
                lost_frames += 1
    else:
        lost_frames += 1
        ball_center = None
    
    # If we recently had a ball, let it "coast" for a bit
    if ball_center is None and prev_ball is not None and lost_frames <= MAX_LOST:
        ball_center = prev_ball
    
    # If we have a valid center now, update prev_ball
    if ball_center is not None:
        prev_ball = ball_center
    else:
        # if we've been lost too long, fully reset
        if lost_frames > MAX_LOST:
            prev_ball = None 

    
    # Draw YOLO boxes FIRST
    annotated = results.plot()
    
    # Now draw the thin rim-top band
    if hoop:
        x1, y1, x2, y2 = hoop
        cv2.rectangle(
            annotated,
            (int(x1), int(y1)),
            (int(x2), int(y2)),
            (0, 255, 255),
            3
        )
    
    # Update state AFTER drawing
    if hoop and ball_center:
        zone = get_ball_zone(ball_center, hoop)
        state = update_state(state, zone)

        bx, by = ball_center
    if state == "shot_started" and not recording:
        trajectory = []
        recording = True
        cooldown = COOLDOWN_FRAMES

    if recording and state != "idle": 
        cooldown = COOLDOWN_FRAMES

    if state == "idle" and recording:
        cooldown -= 1
        if cooldown <= 0:
            recording = False

            # Check if ball was made if interpolation was found
            if x_smooth is not None:
                if did_ball_cross_hoop(hoop, x_smooth, y_smooth):
                    made_shots += 1
                    attempted_shots += 1
                else:
                    if trajectory_near_hoop(hoop, x_smooth, y_smooth):
                        attempted_shots += 1

    frame_idx += 1

    if recording and ball_center:
        bx, by = ball_center
        trajectory.append((frame_idx, bx, by))

    # =======Interpolation of Ball Trajectory=========
    # Interpolate if we have enough points of trajectory
    if recording and len(trajectory) > 5:
        valid = np.array(trajectory)

        frames = valid[:, 0]
        xs     = valid[:, 1]
        ys     = valid[:, 2]
    
        poly_x = np.poly1d(np.polyfit(frames, xs, deg=2))
        poly_y = np.poly1d(np.polyfit(frames, ys, deg=2))
    
        t_smooth = np.linspace(frames.min(), frames.max(), 50)
        x_smooth = poly_x(t_smooth)
        y_smooth = poly_y(t_smooth)
    else:
        x_smooth, y_smooth = None, None


    # =======Drawing of Ball and Text=========
    # Draw state text
    cv2.putText(
        annotated,
        f"State: {state}",
        (30, 50),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 255, 0),
        3
    )

    # Draw points text
    cv2.putText(
        annotated,
        f"Shots Made: {made_shots}/{attempted_shots}",
        (30, 100),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 255, 0),
        3
    )
    
    # Draw ball center
    if ball_center:
        cx, cy = ball_center
        cv2.circle(annotated, (int(cx), int(cy)), 6, (0, 0, 255), -1)

    # Draw shooting line if the ball is shot and interpolating polynomial is available
    if recording and x_smooth is not None:
        for i in range(1, len(x_smooth)):
            cv2.line(
                annotated,
                (int(x_smooth[i-1]), int(y_smooth[i-1])),
                (int(x_smooth[i]),  int(y_smooth[i])),
                (0, 128, 255),
                2
            )


    # Save frame
    out.write(annotated)


cap.release()
out.release()
```
